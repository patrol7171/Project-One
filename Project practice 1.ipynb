```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_df = pd.read_csv("betterdata.csv")
data_df.head()
```

```python
# Remove unnecessary column
del data_df['Unnamed: 0']
data_df.head()
```

```python
# Sort out rows with less than 100 reviews and has "null" for Price
reduced_data_df = data_df.loc[(data_df['Review Count'] > 99)]
reduced_data_df = reduced_data_df.dropna(subset=['Price'])
reduced_data_df.head(20)
```

```python
#Verify counts for each column
reduced_data_df.count()
```

```python
sns.lmplot(x='Country', y='Distance', data=reduced_data_df, fit_reg=False)
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
reduced_data_df.set_index(['Country'], inplace=True)
reduced_data_df.head(20)
```
